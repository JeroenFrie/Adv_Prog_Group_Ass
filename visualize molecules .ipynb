```python
from rdkit import Chem
```

```python
from rdkit.Chem.Draw import IPythonConsole
from rdkit.Chem import Draw
IPythonConsole.ipython_useSVG=True
```

```python
import pandas as pd

def CSV_Loader(filepath):
    csv_df = pd.read_csv(filepath)

    return(csv_df)

Molecule_DF = CSV_Loader("tested_molecules-1.csv")
```

```python
list_mol_1_names = []
list_mol_0_names = []
for molecule in range(len(Molecule_DF)):
    if Molecule_DF['ALDH1_inhibition'][molecule] == 1:
        list_mol_1_names.append(Molecule_DF['SMILES'][molecule])
    else:
        list_mol_0_names.append(Molecule_DF['SMILES'][molecule])
print(list_mol_1_names)

def sort_molecules(list_mol_names):
    list_mol_nr=[]
    list_mol_names.sort(key=len)

    for name in list_mol_names:
        m = Chem.MolFromSmiles(name)
        list_mol_nr.append(m)
    return list_mol_nr
list_mol_0 = sort_molecules(list_mol_0_names)
list_mol_1 = sort_molecules(list_mol_1_names)
```

```python
img_1=Draw.MolsToGridImage(list_mol_1[0:49],molsPerRow=4,subImgSize=(200,200))  
```

```python
img_0=Draw.MolsToGridImage(list_mol_0[0:49],molsPerRow=4,subImgSize=(200,200)) 
```
